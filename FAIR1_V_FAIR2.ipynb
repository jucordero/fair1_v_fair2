```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
T_fair1 = np.load("T_rcp85_fair_1.npy")
T_fair2 = np.load("T_rcp85_fair_2.npy")

C_fair1 = np.load("C_rcp85_fair_1.npy")
C_fair2 = np.load("C_rcp85_fair_2.npy")

F_fair1 = np.load("F_rcp85_fair_1.npy")
F_fair2 = np.load("F_rcp85_fair_2.npy")
```

```python
plt.plot(np.arange(1764, 2500), T_fair1, label = "FAIR 1.6.4")
plt.plot(np.arange(1763, 2500), T_fair2, label = "FAIR 2.0")
plt.xlim(1750, 2100)
plt.xlabel("year")
plt.ylabel("Temperature anomaly [K]")
plt.legend()
```

```python
plt.plot(np.arange(1764, 2500), C_fair1, label = "FAIR 1.6.4")
plt.plot(np.arange(1763, 2500), C_fair2, label = "FAIR 2.0")
plt.xlabel("year")
plt.ylabel("CO2 concentration [ppm]")
plt.legend()
```

```python
plt.plot(np.arange(1764, 2500), F_fair1, label = "FAIR 1.6.4")
plt.plot(np.arange(1763, 2500), F_fair2, label = "FAIR 2.0")
plt.xlabel("year")
plt.ylabel("Radiative Forcing [W/m2]")
plt.legend()
```
